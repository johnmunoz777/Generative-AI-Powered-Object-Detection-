```python
import os

def set_project_dir(project_dir: str):
    """
    Change the current working directory to the given project directory.

    Parameters:
        project_dir (str): Path to the project directory.

    Returns:
        str: The absolute path of the new working directory.
    """
    try:
        os.chdir(project_dir)
        cwd = os.getcwd()
        print(f"âœ… Current working directory set to: {cwd}")
        return cwd
    except Exception as e:
        print(f"âŒ Error: Could not change directory to {project_dir}")
        print("Reason:", e)
        return None


# Example usage:
PROJECT_DIR = "/content/drive/MyDrive/peru_project"
set_project_dir(PROJECT_DIR)
```
